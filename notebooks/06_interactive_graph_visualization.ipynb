# 06 — Interactive Graph Visualization (CholecT50)

**Goal:** Generate interactive, browser-viewable knowledge graphs where:
- **Instrument** nodes are clearly distinguished from **Target** nodes by color and shape
- Node labels show the class type (e.g. "Instrument (grasper)", "Target (gallbladder)")
- Edges show verb name and frequency
- Graphs are fully interactive: drag, zoom, pan, hover for tooltips

**Videos:** VID01, VID05, VID40

### Visual Design
| Node Type | Color | Shape |
|-----------|-------|-------|
| **Instrument** | Blue (#2196F3) | Circle |
| **Target** | Green (#4CAF50) | Diamond |

## Setup & Imports

````python
import sys, os
sys.path.insert(0, os.path.abspath('..'))

import pandas as pd
from pathlib import Path
from src.graph_builder import (
    build_graph,
    build_graphs_for_videos,
    graph_stats,
    print_graph_stats,
    visualize_graph_interactive,
)

# Project paths
PROJECT_ROOT = Path('..').resolve()
TRIPLET_DIR = PROJECT_ROOT / 'outputs' / 'parsed_triplets'
INTERACTIVE_DIR = PROJECT_ROOT / 'outputs' / 'graphs' / 'interactive'
INTERACTIVE_DIR.mkdir(parents=True, exist_ok=True)

print('Parsed CSVs:', sorted(os.listdir(TRIPLET_DIR)))
print(f'Interactive graphs will be saved to: {INTERACTIVE_DIR}')
````

## Build Graphs for All 3 Videos

````python
print('Building graphs for all 3 videos...')
graphs = build_graphs_for_videos(TRIPLET_DIR, video_ids=['VID01', 'VID05', 'VID40'])

for vid, G in graphs.items():
    print_graph_stats(G)
````

## Generate Interactive Graphs

Each graph is saved as a self-contained HTML file that can be opened in any web browser.

**Features:**
- 🔵 **Blue circles** = Instruments (labeled "Instrument (name)")
- 🟢 **Green diamonds** = Targets (labeled "Target (name)")
- Edges show verb + frequency, highlighted orange on hover/click
- Fully interactive: drag nodes, zoom, pan, hover for tooltips
- Legend panel in bottom-left corner

````python
generated_files = []

for vid, G in graphs.items():
    print(f'\n--- {vid} ---')
    print_graph_stats(G)
    
    out_path = INTERACTIVE_DIR / f'{vid}_interactive.html'
    visualize_graph_interactive(G, output_path=out_path)
    generated_files.append(out_path)

print('\n' + '='*60)
print(f'Generated {len(generated_files)} interactive graph(s):')
for p in generated_files:
    print(f'  → {p}')
print(f'\nOpen any HTML file in your browser to explore the graph!')
print('='*60)
````

## Summary

This notebook generates **interactive HTML graphs** for each surgical video where:

1. **Instrument nodes** (e.g., grasper, hook, clipper) are shown as **blue circles**
2. **Target nodes** (e.g., gallbladder, cystic_duct, liver) are shown as **green diamonds**

5. Hover over nodes/edges for **detailed tooltips**
